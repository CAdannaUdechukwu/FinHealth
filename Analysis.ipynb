## DATA PREPARATIONS

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('BCGX_GEN_AI_TASK.csv')
df
```

```python
sort_df = df.sort_values(by=['Company', 'Year'], ascending=[True, True])
sort_df
```

## Year-on-Year Growth (Revenue and Net Income)

```python
sort_df['Revenue Growth (%)'] = sort_df.groupby(['Company'])['Total Revenue'].pct_change() * 100
sort_df['Net Income Growth (%)'] = sort_df.groupby(['Company'])['Net Income'].pct_change() * 100
sort_df
```

```python
sns.lineplot(data=sort_df, x='Year', y='Total Revenue', hue='Company')
plt.xticks(ticks=sort_df['Year'].unique().astype(int))
plt.title('Revenue Over Time (in Millions)')
plt.ylabel('Total Revenue')
plt.xlabel('Year')
plt.show()
```

```python
sns.lineplot(data=sort_df, x='Year', y='Net Income', hue='Company')
plt.xticks(ticks=sort_df['Year'].unique().astype(int))
plt.title('Net Income Over Time (in Millions)')
plt.ylabel('Net Income')
plt.xlabel('Year')
plt.show()
```

## Profitability Analysis

This shows how much each dollar of revenue turns into profit.

```python
sort_df['Net Profit Margin'] = sort_df['Net Income'] / sort_df['Total Revenue']
sort_df
```

```python
sns.scatterplot(data=sort_df, x='Total Revenue', y='Net Income', hue='Company')
plt.title('Revenue vs Net Income (in Millions)')
plt.show()
```

Shows which companies are most efficient at converting revenue into profit

```python
sns.barplot(data=sort_df, x='Company', y='Net Profit Margin')
plt.title('Net Profit Margin by Company (in Millions)')
plt.xticks(rotation=45)
plt.show()
```

## Operating Cash Flow Margin
How much cash the company generates from its core operations

```python
sort_df['Operating Cash Flow Margin'] = sort_df['Cash Flow from OA'] / sort_df['Total Revenue']
sort_df
```

## Debt-to-Asset Ratio
Shows how much of the company is financed by debt

```python
sort_df['Debt-to-Asset-Ratio'] = sort_df['Total Liabilities'] / sort_df['Total Assets']
sort_df
```

```python
plt.xticks(ticks=sort_df['Year'].unique().astype(int))
sns.lineplot(data=sort_df, x='Year', y='Debt-to-Asset-Ratio', hue='Company')
plt.title('Debt-to-Assets Ratio Over Time (in Millions)')
plt.ylabel('Debt Ratio')
plt.show()
```

## Return on Assets (ROA)
Measures how effectively assets are used to generate profit.

```python
sort_df['Return-on-Assets'] = sort_df['Net Income'] / sort_df['Total Assets']
sort_df
```
